```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取数据，跳过前5行
file_path = '/Users/wuzhenhan/Desktop/final-project-xy-wz/data/COVID19_Grant_Report.csv' 
data = pd.read_csv(file_path, skiprows=5)

# 清理 'Award Funding' 列并转换为浮点数
data['Award Funding'] = data['Award Funding'].replace(r'[$,]', '', regex=True).astype(float)

# 移除 'State' 或 'Award Funding' 列中的空值
data = data.dropna(subset=['State', 'Award Funding'])

# 确认所有州名称都是有效的缩写
valid_states = {
    'AL', 'AK', 'AZ', 'AR', 'CA', 'CO', 'CT', 'DE', 'FL', 'GA', 'HI', 'ID', 'IL',
    'IN', 'IA', 'KS', 'KY', 'LA', 'ME', 'MD', 'MA', 'MI', 'MN', 'MS', 'MO', 'MT',
    'NE', 'NV', 'NH', 'NJ', 'NM', 'NY', 'NC', 'ND', 'OH', 'OK', 'OR', 'PA', 'RI',
    'SC', 'SD', 'TN', 'TX', 'UT', 'VT', 'VA', 'WA', 'WV', 'WI', 'WY'
}
data = data[data['State'].isin(valid_states)]

# 计算所有记录的总资助金额（美元）
total_funding_dollars = data['Award Funding'].sum()
print(f"Total Award Funding for all states (Dollars): {total_funding_dollars:,.0f}")

# 将总资助金额转换为百万美元
total_funding_millions = total_funding_dollars / 1e6
print(f"Total Award Funding for all states (Millions $): {total_funding_millions:.2f}")

# 按州分组并计算总资助金额
state_funding = data.groupby('State')['Award Funding'].sum().reset_index()

# 将资助金额转换为百万美元（$M）
state_funding['Award Funding'] = state_funding['Award Funding'] / 1e6  # 单位为百万美元 ($M)

# 按资助金额降序排序
state_funding = state_funding.sort_values(by='Award Funding', ascending=False).reset_index(drop=True)

# 绘制条形图
plt.figure(figsize=(14, 10))
plt.bar(state_funding['State'], state_funding['Award Funding'], color='skyblue')
plt.xlabel('State')
plt.ylabel('Total Award Funding (Millions $)')
plt.title('Total Funding Amount by State')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# 打印资助金额前三的州
print("Top 3 states with the highest funding (in millions):")
print(state_funding.head(3))
```
